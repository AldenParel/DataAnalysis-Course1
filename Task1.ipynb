```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing blackwell data 
data = pd.read_csv(r'C:\Users\alden\OneDrive\Documents\Demographic_Data.csv')
```

```python
#sample data
data.head()
```

```python
#information on data
data.describe()
```

```python
#check data types of each variable in the data
data.info()
```

```python
#clean data and delete duplicates
data = data.drop_duplicates()
```

```python
#check for missing values in data
print(data.isnull().sum())
```

```python
#lists data types of variables
data.dtypes
```

```python
#prints column names
header = data.dtypes.index
print(header)
```

```python
#histogram of instore
plt.hist(data['in-store'])
plt.show()
```

```python
#add bins to histogram
plt.hist(data['in-store'], bins=4)
```

```python
#import numpy libray
import numpy as np

# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)
# printing line plot
fig, ax = plt.subplots()
ax.plot(t, s)
plt.show()
```

```python
#scatter plot of 100 random data samples
data_sample = data.sample(100)

x = data_sample['age']
y = data_sample['amount']
plt.scatter(x,y, marker='o')
plt.show()
```

```python
#creating a box plot
header = data.dtypes.index
print(header)
A = data['amount']
plt.boxplot(A,0,'gD')
plt.show()
```

```python
#creating a correlation matrix
corr_mat = data.corr()
print(corr_mat)
```

```python
#creating a covariance matrix
cov_mat = data.cov()
print(cov_mat)
```
